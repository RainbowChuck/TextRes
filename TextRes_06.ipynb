```python
import asyncio
import streamlit as st

async def async_function():
    await asyncio.sleep(2)
    return "Результат асинхронной функции"

def main():
    st.title("Асинхронный пример")
    
    # Запуск асинхронной функции синхронно
    loop = asyncio.new_event_loop()
    asyncio.set_event_loop(loop)
    result = loop.run_until_complete(async_function())
    
    st.write(f"Результат: {result}")

if __name__ == "__main__":
    main()
import requests
import streamlit as st

def get_vacancies(query):
    response = requests.get(f'https://api.hh.ru/vacancies?text={query}')
    return response.json()

def main():
    st.title("Поиск вакансий")

    # Ввод запроса для поиска
    query = st.text_input("Введите должность")
    
    if query:
        vacancies = get_vacancies(query)
        st.write(f"Найдено {len(vacancies['items'])} вакансий.")
        for vacancy in vacancies['items']:
            st.write(vacancy['name'])
            st.write(vacancy['employer']['name'])
            st.write(vacancy['alternate_url'])
            st.write("---")

if __name__ == "__main__":
    main()
import threading
import time
import streamlit as st

def long_running_task():
    time.sleep(5)
    return "Задача завершена"

def main():
    st.title("Пример с потоками")

    result_placeholder = st.empty()
    
    # Запуск долгой задачи в отдельном потоке
    task_thread = threading.Thread(target=lambda: result_placeholder.write(long_running_task()))
    task_thread.start()

    st.write("Задача выполняется в фоновом потоке...")

if __name__ == "__main__":
    main()
```
